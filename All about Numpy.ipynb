<u> <b> lists vs numpy </u>
1.  Lists are slow whereas numpy arrays are fast. <br>
    Numpy uses fixed type<br>
    Numpy repesents any value as four bytes - int32 by default. <br>
    We can also specify the dtype of elements as int16 or int8 accordignly. less bytes of memory hence faster<br>
    Where as in lists the builtin type is `int` and needs to store following for each value - object value,object type,reference count,size. Hence more size <br>
2. No need of type checking bcoz in numpy everything is ndarray type <br>
3. Numpy uses contiguous memory<br>
   Benefits of contiguous memory - SIMD(Single Instruction Multiple Data) Vector Processing <br>
   Also, effective Cache utilisation
4. Error while performing mathematical operations such as Multiplication of two lists

### Reading an image with OpenCV and creating a greyscale and RGB image with Numpy array

```python
#pip install opencv-python if it is not already present
#cv stands for computer vision
import cv2
img1 = cv2.imread("C:/Users/Vishy Nair/Downloads/OpenCV/cards.jpg")
```

```python
cv2.imshow("Output window",img1)
cv2.waitKey(0)
```

```python
#Again we can see that the type of the image is numpy array
type(img1)
```

```python
#Cropping a part of the iamge
img2 = img1[250:400]
```

```python
cv2.imshow("Output window cropped",img2)
cv2.waitKey(0)
```

```python
#Shape of the Image
img1.shape
```

Here we can see an extra element, for the shape of the ndarray. For that we can try to understand, what is the difference between Greyscale and RGB image.<br>
So the 3 here represents number of channels of the image. A typical greyscale image(or black and white) is something without any channel mentioned. <br>
Number of Channels for a RGB colour image is 3. Hence we need to represent this if we are creating an RGB image <br>

```python
#Creating an RGB image of shape 512 * 512
new_img = np.zeros((512,512,3))
```

```python
new_img.shape
```

```python
#We can see that we have created a RGB image now
cv2.imshow("new window",new_img)
cv2.waitKey(0)
```

```python
new_img[:] = 0,255,0
#For OpenCV colors are in the format BGR. Hence 0,255,0 represents color Green ON
```

```python
cv2.imshow("new window",new_img)
cv2.waitKey(0)
```
