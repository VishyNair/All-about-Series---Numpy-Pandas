#### Dataset Details
##### Meet the Palmer penguins
`CHINSTRAP, ADELIE , GENTOO`<BR>
    
Simplified data from original penguin data sets. 


Contains variables:<br>

- species: penguin species (Chinstrap, Adélie, or Gentoo) <br>
- culmen_length_mm: culmen length (mm)<br>
- culmen_depth_mm: culmen depth (mm)<br>
- flipper_length_mm: flipper length (mm)<br>
- body_mass_g: body mass (g)<br>
- island: island name (Dream, Torgersen, or Biscoe) in the Palmer Archipelago (Antarctica)<br>
- sex: penguin sex<br>
    
Note: The culmen is the upper ridge of a bird’s bill. In the simplified penguins data, culmen length and depth are renamed as variables bill_length_mm and bill_depth_mm

```python
#Writing a simple function to define different levels 'low','medium','high' based on the culmen_length_to_depth value
def culmen_func(x):
    """Culmen Function is to define different levels based on the culmen_length_to_depth. x here woud be culmen_length_to_depth"""
    if (x >=1 and x<2):
        return 'Low'
    elif(x>=2 and x<3):
        return 'Medium'
    elif(x>=3):
        return 'High'
```

```python
#We can apply the above user defined function to the culmen_length_to_depth column to derive a new column
df_new['Culmen_char'] = df['culmen_length_to_depth'].apply(culmen_func)
```

```python
#pivot_table is used to create a spreadsheet-style pivot table as a DataFrame.
#values contains the column to aggregate. If you want all columns, dont mention the values parameter
#If you want more than one column, then pass those features through a list
#index is the row in the pivot table based on which we are grouping by
#columns contains the columns you want in the pivot table
#aggfunc is the aggregate function
df_new.pivot_table(values=['culmen_length_mm','culmen_depth_mm'],index='species',aggfunc='mean')
```

```python
#Another way of using pivot_Table function
df_new.pivot_table(columns='species')
```
